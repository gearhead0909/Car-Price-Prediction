## Importing Libraries

```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import ExtraTreesRegressor
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
from sklearn.metrics import r2_score
```

## Loading Data

```python
Dataset = pd.read_csv("cars_price.csv")
```

```python
Dataset.columns
```

```python
Dataset.head()
```

```python
Dataset = Dataset.dropna()
Dataset = Dataset.drop(['Unnamed: 0'], axis=1)
```

```python
Dataset.head()
```

```python
Dataset.shape
```

## Data pre-processing

```python
Data = Dataset.copy()
```

```python
Data = Data.drop(["model"], axis = 1)
```

```python
Data.head()
```

```python
# Categorical Variables are converted into Numerical Variables with the help of Dummy Variable
```

```python
Condition    = pd.get_dummies(Data['condition'], drop_first = True)
Fuel_Type    = pd.get_dummies(Data['fuel_type'], drop_first = True)
Color        = pd.get_dummies(Data['color'], drop_first = True)
Transmission = pd.get_dummies(Data['transmission'], drop_first = True)
Drive_unit   = pd.get_dummies(Data['drive_unit'], drop_first = True)
Segmenmt     = pd.get_dummies(Data['segment'], drop_first = True)
```

```python
# Removing old variables
```

```python
Data = Data.drop(['make', 'condition', 'fuel_type', 'color', 'transmission', 'drive_unit', 'segment'], axis = 1)
```

```python
Data.head()
```

```python
## Adding New obtained Data
```

```python
Data = pd.concat([Data, Condition, Fuel_Type, Color, Transmission, Drive_unit, Segmenmt], axis = 1)
```

```python
Data.head()
```

## Splitting the Data into Training and Testing sets

```python
Data_train, Data_test = train_test_split(Data, train_size = 0.7, test_size = 0.3, random_state = 100)
```

## Noarmalizing Numerical Data

```python
scaler = MinMaxScaler()
num_vars = ['priceUSD', 'year', 'mileage(kilometers)', 'volume(cm3)']

x = Data_train[num_vars].values
x_scaled = scaler.fit_transform(x)
Data_train[num_vars] = x_scaled

x = Data_test[num_vars].values
x_scaled = scaler.fit_transform(x)
Data_test[num_vars] = x_scaled
```

```python
Data_train.head()
```

## Calculating Correlation coefficents

```python
plt.figure(figsize = (16, 10))
sns.heatmap(Data_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

## Splitting Data and Labels

```python
y_train = Data_train.pop('priceUSD')
X_train = Data_train
```

```python
model = ExtraTreesRegressor()                        # Making the model
rfe = RFE(model, n_features_to_select=10)            # Number of features to be selected
fit = rfe.fit(X_train, y_train)                     # Training the data

list(zip(X_train.columns, fit.support_, fit.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

## Building model using statsmodel, for the detailed statistics

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
print(lm.summary())
```

```python
vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Residual Analysis of the train data

```python
y_train_price = lm.predict(X_train_rfe)
```

```python
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```

## Making predictions

```python
y_test = Data_test.pop('priceUSD')
X_test = Data_test[col]
```

```python
X_test_rfe = sm.add_constant(X_test)
```

```python
y_pred = lm.predict(X_test_rfe)
```

## Model Evalution

```python
r2_score(y_test, y_pred)
```

```python
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred')                           # Plot heading 
plt.xlabel('y_test')                                       # X-label
plt.ylabel('y_pred')                                       # Y-label
```
